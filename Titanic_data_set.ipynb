```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")
print("First 5 Rows:")
print(data.head())
```

```python
print("\nDataset Info:")
print(data.info())
print("\nStatistical Summary:")
print(data.describe())
```

```python
print("\nMissing Values:")
print(data.isnull().sum())

data["Age"] = data["Age"].fillna(data["Age"].mean())
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])

print("\nMissing Values After Filling:")
print(data.isnull().sum())
```

```python
print("\nFirst Passenger Details (loc):")
print(data.loc[0])

print("\nFirst 5 Passengers (iloc):")
print(data.iloc[0:5])

print("\nSelected Columns (Age, Fare, Survived):")
print(data[["Age", "Fare", "Survived"]].head())
```

```python
data.rename(columns={"Sex": "Gender"}, inplace=True)
print("\nAfter Renaming Sex to Gender:")
print(data.head())
```

```python
print("\nPassengers Age > 30:")
print(data[data["Age"] > 30].head())

print("\nPassengers Fare > 50:")
print(data[data["Fare"] > 50].head())

print("\nFemale Passengers Who Survived:")
print(data[(data["Gender"] == "female") & (data["Survived"] == 1)].head())
```

```python
print("\nSorted by Fare (Highest First):")
print(data.sort_values(by="Fare", ascending=False).head())

print("\nSorted by Age:")
print(data.sort_values(by="Age").head())
```

```python
print("\nSurvival Rate by Gender:")
print(data.groupby("Gender")["Survived"].mean())


print("\nAverage Age by Pclass:")
print(data.groupby("Pclass")["Age"].mean())

print("\nSurvival Count by Embarked:")
print(data.groupby("Embarked")["Survived"].sum())
```

```python
print("\nAverage Age by Survival Status:")
print(data.groupby("Survived")["Age"].mean())
```

```python
class_info = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Type": ["First Class", "Second Class", "Third Class"]
})

merged_data = pd.merge(data, class_info, on="Pclass")

print("\nMerged Data:")
print(merged_data.head())

data_part1 = data.iloc[:100]
data_part2 = data.iloc[100:200]

concat_data = pd.concat([data_part1, data_part2])
print(concat_data.shape)
```
